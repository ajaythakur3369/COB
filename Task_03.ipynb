```python
plt.figure(figsize = (15, 6))
sns.countplot(x = new_data_2['country'], order = new_data_2['country'].value_counts().index[0:10], width = 0.5)
plt.grid(True, color = "grey", linewidth = "0.4")
plt.xlabel('country', fontname = 'serif', fontsize = 14)
plt.xticks(fontname = 'serif')
plt.ylabel('count', fontname = 'serif', fontsize = 14)
plt.yticks(fontname = 'serif')
plt.title("Country Wise Content On Netflix", fontname = 'serif', fontsize = 16)
plt.show()
```
