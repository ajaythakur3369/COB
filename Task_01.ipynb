# **Problem Statement -**

Create a CSV dataset using Python, Pandas, and any Public API.

**Dataset -** hhttps://www.binance.com/api/v3/ticker/24hr

# **Let's Begin !**

```python
# Import all required Libraries
import pandas as pd
import requests
import warnings
warnings.filterwarnings('ignore')
```

```python
# Call API and data through API
api_url = "https://www.binance.com/api/v3/ticker/24hr"
```

```python
response = requests.get(api_url)
```

```python
# Access the Drive from Colab, to access the file
from google.colab import drive
drive.mount('/content/drive/')
```

```python
# Put If-else statement on called API
if response.status_code == 200:
    data = response.json()
    df = pd.DataFrame(data)
    # This will create a CSV file named "Binance_dataset.csv" in the current directory
    df.to_csv('/content/drive/MyDrive/Folders/Binance_dataset.csv', index = False)
    print("Your CSV dataset has been created successfully.")
else:
  print("Unable to create CSV file. Check your url")
```
